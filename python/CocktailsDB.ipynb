```python
import pandas as pd
import requests as rs
```

```python
liquors = ['Vodka', 'Gin', 'Tequila', 'Rum', 'Light_Rum', 'Dark_Rum', 'Brandy', 'Whiskey', 'Blended_Whiskey', 'Bourbon', 'Sweet_Vermouth', 'Dry_Vermouth', 'Scotch']
df_list = []

for search_term in liquors:
    cocktail_url = f'https://www.thecocktaildb.com/api/json/v1/1/filter.php?i={search_term}'
    results = rs.get(cocktail_url).json()
    df_list.append(results['drinks'])
#     df = pd.DataFrame(results['drinks'])
#     df['base_liquor'] = search_term
#     df_list.append(df)
#     db.cocktails.insert_many(df.to_dict('records'))
```

```python
filtered_list = []
for i in range(len(df_list)):
    for j in range(len(df_list[i])):
        temp = df_list[i][j]
        filtered_list.append(temp)
```

```python
filtered_list
```

```python
ids = []

for i in range(len(filtered_list)):
    id = filtered_list[i]['idDrink']
    ids.append(id)
```

```python
details = []
for id in ids:
    url = f'https://www.thecocktaildb.com/api/json/v1/1/lookup.php?i={id}'
    results = rs.get(url).json()
    details.append(results)
```

```python
details_list = []

for i in range(len(details)):
    temp = details[i]['drinks'][0]
    details_list.append(temp)
```

```python
details_list
```

```python
temp_1 = pd.DataFrame(filtered_list)
temp_2 = pd.DataFrame(details_list)
```

```python
df_merged = temp_1.merge(temp_2, on=['idDrink', 'strDrink', 'strDrinkThumb'], suffixes=("", ""))

del df_merged['strDrinkThumb']

df_merged.head()
```

```python
df_merged.drop(['strDrinkAlternate', 'strDrinkES', 'strDrinkDE',
                'strDrinkFR', 'strDrinkZH-HANS', 'strDrinkZH-HANT', 
                'strTags', 'strVideo', 'strIBA', 'strGlass', 
                'strInstructionsES', 'strInstructionsDE', 'strInstructionsFR', 
                'strInstructionsZH-HANS', 'strInstructionsZH-HANT', 'strCreativeCommonsConfirmed', 'dateModified'], axis=1, inplace=True)

df_merged.columns
```

```python
df_merged.to_csv("Cocktails.csv", index=False)
```
